```python
from rdflib.graph import ConjunctiveGraph as Graph
from rdflib.store import Store
from rdflib.plugin import get as plugin
from rdflib.term import URIRef
from rdflib import OWL, RDFS, Literal, Namespace
from rdflib.namespace import FOAF, RDF, XSD
import csv, os, time, shutil
from progress1 import Progress
from utils import uri_base, is_specified

import pandas as pd


identifier = 'stopsGraph'
filename = 'stops.csv'
batches = 100

g = Graph(identifier=identifier)
n = Namespace(uri_base)

total_entries = num_lines = sum(1 for line in open(filename, encoding="utf8")) - 1

progress = Progress(total_entries, batches)
with open(filename, encoding="utf8") as fd:
    data0 = pd.read_csv(r'.\stops.csv')
    data = data0.T
    
    for row in data:

        stop_node = n['Stops/UrbanStops/' + str(data[row]['stop_id'])]
        class_node = n['UrbanStops']
        
        g.add((stop_node, n.stop_id, Literal(data[row]['stop_id'])))
        g.add((stop_node, n.stop_code, Literal(data[row]['stop_code'])))
        g.add((stop_node, n.stop_name, Literal(data[row]['stop_name'])))
        g.add((stop_node, n.stop_desc, Literal(data[row]['stop_desc'])))
        g.add((stop_node, n.location_type, Literal(data[row]['location_type'])))
        
        
        g.add((stop_node, n.stop_lat, Literal('point(' + str(data[row]['stop_lat']) + ' ' + str(data[row]['stop_lon']) + ')^^ <http://www.openlinksw.com/schemas/virtrdf#Geometry>')))
        
        g.add((stop_node, RDF.type, class_node))
        

        progress.count()
        if progress.is_batch_complete(): 
            output = 'outs/stops/' + identifier + '{num:0{width}}'.format(num=progress.current_batch, width=4) + '.ttl.n3'
            g.serialize(destination=output, format='turtle')
            g.close()
            g = Graph(identifier=identifier)

        if progress.finished():
            break
g.close()
```
